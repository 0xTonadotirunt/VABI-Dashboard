```python
import pandas as pd

```

```python
df = pd.read_csv('data/owid.csv')
df.head()
```

```python
cols = [
    'gdp',
    'nitrous_oxide_per_capita',
    'methane_per_capita',
    'gas_co2_per_capita',
    'total_ghg',
    'country',
    'co2_per_capita',
    'year',
    'iso_code',
    'population',
    'gdp',
    'temperature_change_from_ch4',
    'temperature_change_from_ghg',
    'temperature_change_from_co2',
    'temperature_change_from_n2o',
    'land_use_change_co2_per_capita',
    'co2',
    'methane',
    'nitrous_oxide',
    'total_ghg'
]

trunc_df = df[cols]




 
```

```python
df_ghg = pd.read_csv('data/owid_ghg.csv')

df_ghg.columns
```

```python
df_ghg_melted = pd.melt(df_ghg, id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code' ], var_name="year", value_name='emissions')

df_ghg_melted_dropped_na = df_ghg_melted.dropna(subset=['emissions'])
df_ghg_melted_dropped_na.head()
df_ghg_melted_dropped_na['year'] = df_ghg_melted_dropped_na['year'].astype(int)
df_ghg_melted_dropped_na['Country Code'] = df_ghg_melted_dropped_na['Country Code'].astype(str)
```

```python
merged_df = pd.merge(trunc_df, df_ghg_melted_dropped_na, left_on=['iso_code', 'year'], right_on=['Country Code', 'year'])
```

```python
merged_df.head()

merged_df.rename(columns={"emissions": "HFC_PFC_others"})
```

```python
merged_df.to_csv('data/emissions_merged_no2_ch4_.csv', index=False)
```
